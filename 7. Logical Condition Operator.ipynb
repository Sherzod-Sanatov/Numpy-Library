1. Replace Zeros: Create an array with random values between -5 and 5. Replace all zeros with the value 100 using np.where.

```python
import numpy as np
```

```python
cond=([True,True,True,True,True,False,True,True,True,True])
```

```python
y=100
```

```python
x=np.arange(-5,5)
x
```

```python
arr1=np.where(cond,x,y)
arr1
```

2. Greater Than Average: Create an array with random values. Find the average and use np.where to create a new array containing only values greater than the average.

```python
x=np.random.randn(3,3)
x
```

```python
arr=np.mean(x)
arr
```

```python
arr2=np.where(x>arr,x,0)
arr2
```

3. Alternating Values: Create two arrays of equal length. Use np.where to create a new array that alternates between elements from the two arrays.

```python
x=np.arange(6)
y=np.arange(6,12)
```

```python
c=([True, False, True, False, True, False])
```

```python
a=np.where(c,x,y)
a
```
