Exercises for np.unique:

Create an array with various repeated elements. Find the unique elements and their counts.


```python
import numpy as np
```

```python
arr=np.array([3,33,33,44,44,234,23,4,66,4,234])
```

```python
uq_elem, count=np.unique(arr, return_counts=True)
print(uq_elem)
count
```

Exercises for np.setdiff1d:

Create two arrays representing grocery lists. Find the items that are only in the first list.
Generate two arrays of random numbers. Identify the numbers that exist in the first array but not in the second.

```python
arr1=np.array(['cookies','donuts','sprite','chips','milk'])
arr2=np.array(['lamb','stake','milk','water','donuts','chips'])
print(arr1,arr2)
```

```python
diff=np.setdiff1d(arr1,arr2)
diff
```

```python
arr1=np.random.randint(10,30,10)
arr1
```

```python
arr2=np.random.randint(10,30,10)
arr2
```

```python
diff=np.setdiff1d(arr1,arr2)
diff
```

Exercises for np.in1d:

Create an array of student grades and a separate array of passing grades. Check which students passed.
Generate a list of email addresses and a list of subscribed users. Verify which emails belong to subscribers.

```python
grades=np.array([3,4,2,5,1,3,3,4,5,3,2,4,5])
passing_grades=np.array([4,5])
```

```python
results=np.in1d(grades,passing_grades)
results
```

```python
emails=np.array(['sher.san@bk.com','suxa.ochilov@02.ru','jama.jamti@2000.com'])
subscribers=np.array(['sher.san@bk.com', 'axa.ochilov@97.uz'])
```

```python
results=np.in1d(emails,subscribers)
results
```
