1. Create a 3x4 array filled with random integers between 1 and 10. Calculate the mean of all elements, the mean of each row, and the mean of each column.

```python
import numpy as np
```

```python
arr=np.random.randint(1,10,10).reshape(2,5)
arr
```

```python
mean=np.mean(arr)
mean
```

```python
mean_row=np.mean(arr,axis=1)
mean_row
```

```python
mean_column=np.mean(arr,axis=0)
mean_column
```

2. Generate a 5x5 array with values increasing linearly from 1 to 25. Calculate the cumulative sum along each row and each column.

```python
arr=np.arange(25).reshape(5,5)
arr
```

```python
cumsum=np.cumsum(arr)
cumsum
```

```python
cumsum_1=np.cumsum(arr, axis=1)
cumsum_1
```

```python
cumsum_0=np.cumsum(arr,axis=0)
cumsum_0
```
