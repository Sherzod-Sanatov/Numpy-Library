```python
import numpy as np
```

One Dimensional Indexing

```python
arr1d=np.arange(5,26,5)
arr1d[1]
```

```python
arr=np.arange(2,11,2)
arr[-1]
```

```python
arr=np.arange(1,10,2)
arr[3]=10
arr
```

Slicing

```python
arr=np.arange(3,28,3)
arr[2:7]
```

```python
arr=np.arange(2,19,2)
arr[:5]
```

```python
arr = np.array([1, 4, 9, 16, 25, 36, 49, 64, 81])
arr[::2]
```

2 Dimensional indexing

```python
arr_2d = np.array([[3, 6, 9], [12, 15, 18], [21, 24, 27]])
arr_2d[1:,0:2]
```

```python
arr_2d = np.array([[2, 4, 6], [8, 10, 12], [14, 16, 18]])
arr_2d[0,:2]
```
