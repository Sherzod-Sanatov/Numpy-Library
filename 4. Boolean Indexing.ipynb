1. Greater Than Threshold:

Given an array arr = np.array([10, 15, 20, 25, 30]), use boolean indexing to print elements greater than 20.

```python
arr=np.arange(10,31,5)
arr[arr>20]
```

2. Even Numbers:

Create an array arr = np.array([2, 7, 12, 18, 23, 30]). Use boolean indexing to print only the even numbers.

```python
arr = np.array([2, 7, 12, 18, 23, 30])
arr[arr%2==0]
```

3. Negative Values:

Given the array arr = np.array([-3, 5, -8, 10, -12, 15]), use boolean indexing to print only the negative values.

```python
arr = np.array([-3, 5, -8, 10, -12, 15])
arr[arr<0]
```

4. Divisible by 3:

Create an array arr = np.array([9, 14, 21, 26, 33, 38]). Use boolean indexing to print elements divisible by 3.

```python
arr = np.array([9, 14, 21, 26, 33, 38])
arr[arr%3==0]
```

5. Between a Range:

Given the array arr = np.array([25, 30, 35, 40, 45, 50]), use boolean indexing to print elements between 30 and 45 (inclusive).

```python
arr = np.array([25, 30, 35, 40, 45, 50])
```

```python
arr[(arr>=30) & (arr<=45)]
```
